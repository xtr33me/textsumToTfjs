```python
import tensorflow as tf
import os
import numpy as np
import os,glob,cv2
import sys,argparse
from tensorflow.python.framework import graph_util
from tensorflow.python.platform import gfile
```

```python
input_model_path = './saved_model/model.ckpt-178311'
output_graph='./saved_model/model.pb'
```

## Restore model

```python
saver = tf.train.import_meta_graph(input_model_path + '.meta', clear_devices=True)
graph = tf.get_default_graph()

sess = tf.Session()
saver.restore(sess, input_model_path)
input_graph_def = sess.graph.as_graph_def()
```

## Choose outputs

```python
output_node_names=[n.name for n in tf.get_default_graph().as_graph_def().node]
```

```python
[print(n.name) for n in tf.get_default_graph().as_graph_def().node]
```

```python
output_graph_def = graph_util.convert_variables_to_constants(
            sess, # The session
            input_graph_def, # input_graph_def is useful for retrieving the nodes 
            output_node_names  
)
```

## Serialize

```python

with tf.gfile.GFile(output_graph, "wb") as f:
    f.write(output_graph_def.SerializeToString())

LOGDIR='./tensorboard'
train_writer = tf.summary.FileWriter(LOGDIR)
train_writer.add_graph(sess.graph)
sess.close()
```

```python
with tf.gfile.GFile(output_graph, "rb") as f:
    restored_graph_def = tf.GraphDef()
    restored_graph_def.ParseFromString(f.read())
```

```python
with tf.Graph().as_default() as graph:
    tf.import_graph_def(
        restored_graph_def,
        input_map=None,
        return_elements=None,
        name=""
        )
```
